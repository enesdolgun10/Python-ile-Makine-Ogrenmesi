# NumPy ile Matrislerde Dört İşlem

#### NumPy kullanarak toplama, çıkarma, çarpma ve bölme işlemleri

```python
import numpy as np
```

```python
# örnek matrislerimiz
```

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])
print(A.shape, "  ", B.shape)
```

```python
# Toplama (A+B)
```

```python
toplam = A + B
print(toplam)
```

```python
# Çıkarma (A-B)
```

```python
cikarma = A - B
print(cikarma)
```

```python
# Çarpma
```

```python
# 1) Eleman Bazlı Çarpma : A * B
```

```python
carpma_eleman_bazli = A * B
print(carpma_eleman_bazli)
```

```python
# 2) Matris Çarpımı (Lineer Cebir) : A @ B veya np.dot(A,B) 
```
