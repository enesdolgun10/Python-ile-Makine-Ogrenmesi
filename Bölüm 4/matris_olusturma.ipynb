# NumPy Matris Oluşturma Örnekleri

```python
import numpy as np
```

```python
# 1) Tek boyutlu array (vektör gibi)
a = np.array([5, 7, 3, 4]) # 1 boyutlu array (4 eleman)
print(a)
a.shape
```

```python
# 2) İki boyutlu array   1 satır 4 sutun
b = np.array([[5,7,4,3]])
print(b)
b.shape
```

**Yukarıdaki örnekte dikkat! İkisinde de içerideki liste aynı ama birinde tek boyutlu diğerinde 2 boyutlu nump array oluşuyor!**

```python
# 3) İki boyutlu array  2 satır 1 sütun
c = np.array([[7],[8]])
print(c)
c.shape
```

```python
# 4) İki boyutlu array   2x2 matris
d = np.array([[1,2],[3,4]])
print(d)
d.shape
```

```python
# 5) İki boyutlu array   2x3 matris
e = np.array([[1,2,3],[4,5,6]])
print(e)
e.shape
```

```python
# 6) Birim matris (eye)
I = np.eye(4)
print(I)
```

```python
# 7) Sıfır ve birlerle matris
zeros = np.zeros((2,3))
print(zeros)
```

```python
ones = np.ones((3,2))
print(ones)
```

```python
# 8) Random matris
rand_m = np.random.randint(50,100,(2,3))   # 2x3 rastgele tam sayılarla matris
print(rand_m)
```
