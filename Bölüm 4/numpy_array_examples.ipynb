# NumPy Array Oluşturma Örnekleri

NumPy'da array oluşturmanın birden fazla yöntemi mevcuttur. Ancak kendi belirlediğiniz spesifik değerler içeren özel bir numpy array oluşturabilmeniz için mutlaka ya bir list ya da tuple kullanmanız gerekmektedir.

```python
import numpy as np
```

```python
# 1) Liste veya tupledan array oluşturma
arr1 = np.array([1, 3, 87, 44])  # listeden
```

```python
type(arr1)
```

```python
arr2 = np.array((10, 20, 30)) # tupledan
```

```python
type(arr2)
```

```python
print(arr1)
print(arr2)
```

```python
# 2) Çok boyutlu array
arr3 = np.array([[1,2,3],[4,5,6]]) # 2x3 array
print(arr3)
```

```python
type(arr3)
```

```python
arr3.shape
```

```python
listx = [[1,2,3],[4,5,6]]
```

```python
listy = [[10,20,30],[40,50,60]]
```

```python
listz = listx + listy
print(listz)
```

```python
arr4 = np.array([[10,20,30],[40,50,60]])
```

```python
arr5 = arr4 + arr3
print(arr5)
```

```python
print(arr4)
```

```python
# 3) Sıfır, bir değerleriyle dolu ve boş array oluşturma
```

```python
z = np.zeros((2,3)) # 2x3 sıfırlar
print(z)
```

```python
o = np.ones((3,2))  # 3x2 birler
print(o)
```

```python
e = np.empty((2,2))  # 2x2 boş (rastgele değerler gelebilir)
print(e)
```

```python
e[0][0]= 40
print(e)
```

```python
z[1,1] = 5
print(z)
```

```python
w = np.zeros(5) # varsayılan olarak dtype = float olur yani 0.0
print(w)
```

```python
w[0] = 8
print(w)
type(w[0])
```

```python
w2 = np.zeros(5,dtype=int)
print(w2)
type(w2[0])
```

```python
w2[0] = 9
print(w2)
```

```python
# 4) Sayı aralıklarıyla array
a1 = np.arange(0, 10, 2) # 0'dan 10'a  kadar 2şer
print(a1)
```

```python
s2 = np.linspace(0, 1, 5) # 0 ile 1 arasındaki 5 sayı
print(s2)
```

```python
# 5) Rastgele (Random) array
r1 = np.random.rand(2,3) # 2x3, 0-1 arası float değerlerle
print(r1)
```

```python
r2 = np.random.randint(0,10,(2,3))  # 2x3 luk 0-9 arası integer 
print(r2)
```

```python
# 6) dtype (veri tipi) ile array
arr4 = np.array([4.3, 7, 3],dtype=float)  # float tipinde
print(arr4)
```

```python
arr5 = np.array([4, 7, 3],dtype=int) # int tipinde
print(arr5)
```

```python
arr_bool = np.array([1, 0, 1],dtype=bool)  # bool tipinde
print(arr_bool)
```

```python
# 7) dtype kullanmazsanız
arr7 = np.array([1, 2, 3])  # int yapar dtype vermezsek
print(arr7,type(arr7[0]))
```
